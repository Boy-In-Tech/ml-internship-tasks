```python
import pandas as pd
house_prices = pd.read_csv("housePredictionDataSet.csv", sep="\s+", header=None)
# house_prices
```

```python
house_prices.columns = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
    "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"
]
house_prices.head()
```

```python
"""I began by loading the 4)House prediction data set dataset into a pandas DataFrame.
Then, I assigned descriptive column names(Using the classical boston dataframe as inspiration) so each feature could be easily identified."""
```

```python
house_prices.isnull().sum()
```

```python
"""I checked for missing data to ensure the dataset is complete.
In this case, there are no missing values, which means I can continue without filling or dropping data."""
```

```python
X = house_prices.drop("MEDV", axis=1)
y = house_prices["MEDV"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
"""I standardised the features so that they share a similar scale.
This prevents features with larger numeric values (like TAX) from dominating smaller ones (like NOX)."""
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print("Preprocessing complete!")
print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```
